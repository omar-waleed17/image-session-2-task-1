```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
def display_img(img):
    new_img=cv2.cvtColor(img,cv2.COLOR_RGB2BGR)
    plt.imshow(new_img)
    plt.show()
```

```python
def sobel(img):
 gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

 G_blur=cv2.GaussianBlur(gray,(3,3),100)
 Sobel_X=cv2.Sobel(G_blur,-1,1,0) 
 Sobel_Y=cv2.Sobel(G_blur,-1,0,1)
 Sobel_or=cv2.bitwise_or(Sobel_X,Sobel_Y)
 return Sobel_or
```

```python
def canny(img):
 gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
 blur=cv2.GaussianBlur(gray,(3,3),100)
 canny_image=cv2.Canny(blur,22,80)
 return canny_image
```

```python
def laplacian(img):
    gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    laplacian_img = cv2.Laplacian(gray_img, -1)
    return laplacian_img
```

```python
img=cv2.imread("images.jpg")
plt.title("Original")
display_img(img)
sobel_image=sobel(img)
plt.title("Sobel")
display_img(sobel_image)
plt.title("canny")
canny_image = canny(img)
display_img(canny_image)
laplacian_image = laplacian(img)
plt.title("laplacian")
display_img(laplacian_image)
```
